### Observations of the datasets

The HistoricalRateDetail dataset was imported with metadata included in the first two rows. The first row contains a description of the exchange-rate series, while the second row contains the actual variable names ("Date" and "Value"). These rows will need to be excluded during the data cleaning stage before analysis.

- The dataset contains 2,843 rows and 2 columns.
- The first row describes the exchange-rate series.
- The second row contains the actual headers ("Date" and "Value").
- The exchange-rate observations appear to be daily.
- The data is from 02 January 2015 to 18 May 2026.
- The variable of the data is Rand per US Dollar ( USD/ZAR exchange rate).
- The exchange rates are ordered from the most recent to the oldest date.
- The dataset will require restructuring before it can be merged with the CPI datasets.
- The dataset is from SARB.


#### EXCEL - CPI (COICOP 2018 - 8digit) (202604)
- The dataset contains 391 observations and 233 variabes.
- The variables consist of 221 floating-point, 6 integer, and 6 string data types.
- The first seven columns contain classification information, including Division, Group, Class, Subclass, product codes, and CPI weights.
- The dataset is structured hierarchically using the COICOP 2018 classification system, consisting of 13 divisions, 44 groups, and numerous detailed subclasses.
- The DivisionDescription variable identifies broad expenditure categories, while GroupDescription, ClassDescription, and SubclassDescription provide increasingly detailed classifications.
- Monthly CPI index values are recorded from January 2008 (M200801) to April 2026 (M202604).
- The dataset includes detailed food-related subclasses such as cereals, flour of cereals, bread and bakery products, meat, and fish, which align directly with the focus of this research.
- The dataset originates from Statistics South Africa (Stats SA) and provides sufficient historical coverage for the study period.
- The dataset contains all CPI divisions but the study is specifically about food. One of the first preprocessing steps will be to filtwr the dataset to that.


#### CPI_Average prices_Provinces(202604)

- The dataset contains 2,455 observations and 124 variables.
- Monthly average retail prices are reported from January 2017 through April 2026.
- Unlike the COICOP dataset, this dataset records actual prices in South African Rand rather than CPI index values.
- Each observation represents a specific product measured using a defined unit (for example, White Rice – 1 Kilogram).
- Several observations contain missing monthly values, indicating incomplete price records for some products or provinces.
- The dataset appears to contain repeated products, suggesting that prices are recorded separately across provinces.
- The dataset originates from Statistics South Africa.

#### CPI_Average Prices_All urban(202604)
- The dataset contains 391 observations and 123 variables.
- It records monthly average retail prices for a wide range of consumer products from January 2017 (M201701) to April 2026 (M202604).
- Similar to the CPI_Average prices_Provinces dataset, this dataset contains actual average retail prices in South African Rand.
- Each observation represents a specific product together with its unit of measurement (e.g., White Rice – 1 Kilogram, Cake Flour – 2.5 Kilograms).
- Some variables contain missing values, particularly the Old code field and certain monthly observations, indicating that not all products have complete historical records.
- The dataset represents national average (All Urban) prices, whereas the provincial dataset provides prices at the provincial level.
- The dataset originates from Statistics South Africa (Stats SA).


Four datasets were examined for their relevance to the research objectives. The SARB HistoricalRateDetail dataset provides daily USD/ZAR exchange rates, while the COICOP 2018 (8-digit) dataset provides monthly CPI indices across detailed food categories. These two datasets were selected as the primary datasets because they directly support the modelling of exchange rate pass-through into food prices. The Average Prices (All Urban) and Average Prices (Provinces) datasets provide complementary information on retail prices and may be used for descriptive analysis.

| Dataset                         | Frequency | Measures                       | Research Role                  |
| ------------------------------- | --------- | ------------------------------ | ------------------------------ |
| **HistoricalRateDetail (SARB)** | Daily     | USD/ZAR exchange rate          | Independent variable           |
| **COICOP 2018 (8-digit)**       | Monthly   | CPI index by food category     | **Primary dependent variable** |
| **Average Prices (All Urban)**  | Monthly   | National average retail prices | Supporting dataset             |
| **Average Prices (Provinces)**  | Monthly   | Provincial retail prices       | Extension dataset   |
